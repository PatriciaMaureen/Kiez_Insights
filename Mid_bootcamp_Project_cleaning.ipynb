```python
import pandas as pd
import numpy as np

pd.set_option('display.max_columns', None)
```

```python
df = pd.read_csv('giessdenkiez_bewässerungsdaten.csv')
```

```python
df
```

```python
#renaming columns
df = df.rename(columns={'bezirk':'district','baumart':'species','gattung':'genus','strname':'street', 'pflanzjahr':'year','zeitpunkt_der_bewaesserung':'timing_of_watering','bewaesserungsmenge_in_liter':'watering_quantity_in_liters'})
```

```python
df.head(50)
```

```python
#There are some rows where the data is shifted by one column. We want to fix this first, before we are looking at any further cleaning.

# Identify rows with misplaced data in the year column.
misplaced_rows = df[df['x'].notna()]

# Iterate over misplaced rows to realign the data
for index, row in misplaced_rows.iterrows():
    # Extract misplaced data from the 'district' column
    misplaced_data = row['district']
    
    # Move the misplaced data to the correct column
    df.at[index, 'genus'] = row['street']
    df.at[index, 'street'] = row['year']
    df.at[index, 'year'] = row['timing_of_watering']
    df.at[index, 'timing_of_watering'] = row['watering_quantity_in_liters']
    df.at[index, 'watering_quantity_in_liters'] = row['x']
    
```

```python
df.head(50)
```

```python
# It worked so now we can drop column x.
df = df.drop(['x'], axis=1)
df
```

```python
# Check the types of the columns

df.dtypes
```

```python
# I want to change year to an int, time_of_watering to date and watering_quantity_in_liters to numeric
df['year'] =  pd.to_numeric(df['year'], errors='coerce').astype('Int64')
df['watering_quantity_in_liters'] =  pd.to_numeric(df['watering_quantity_in_liters'], errors='coerce')
```

```python
df.dtypes
```

```python
# Check for NaN values
df.isna().sum()
```

```python
df[df['genus'].isna()]['species'].unique()
```

```python
# As we can see the species is quite unique so for my analysis I would rather keep the genus. 
#If we have a look at the categories in genus we can see that there are a lot of different categories so in the further steps I will not keep all of them.
#Therefore I now will fill the missing genus with other and later I can use the category to group more of them together.

df['genus'].value_counts()
```

```python
df['genus'].fillna('other', inplace=True)
df.isna().sum()
```

```python
df[df['street'].isna()==True].head(20)
```

```python
df['street'].value_counts()
```

```python
# Because we do have all the other information, it does not seem that the data is missing. 
# Since there already is a category called "undefined" I would add the NaN values there

df['street'].fillna('undefined', inplace=True)
df.isna().sum()
```

```python
#For this trees we do not know how old they are but I do not want to drop the rows. So I will calculate the average year for the other trees and take the mean as a replacement.
df[df['year'].isna()==True].head(20)
```

```python
round(df['year'].mean())
```

```python
df['year'] = df['year'].fillna(round(df['year'].mean()))
df.isna().sum()
```

```python
df[df['timing_of_watering'].isna()==True].head(20)
```

```python
# There is no good replacement for timin_of_watering and if I want to look at correlations between the weather I need the timing.
# So I will drop these 10 rows.

df = df.dropna(subset=['timing_of_watering'])
df
```

```python
# No I want to check for duplicates
df.duplicated().sum()
```

```python
# Filter the DataFrame to show duplicate rows
df[df.duplicated()]
```

```python
# As we can see that are no real duplicates. So we can keep them in our dataset.
```

```python
#Let's check our weather data
df_weather = pd.read_csv('weather_data_2020-2022.csv')
df_weather = df_weather.drop(['Unnamed: 0'], axis=1)
df_weather
```

```python
df_weather.dtypes
```

```python
df_weather.isna().sum()
```

```python
df_weather.duplicated().sum()
```

```python
# It looks like we can directly work with this data. I want to merge the two dataframes based on the date.
# The weather data is a daily average so I do not need the time but the date in both datasets.
# First I need to convert date to datetime

df_weather['date'] = pd.to_datetime(df_weather['date'])
df_weather['date'] = df_weather['date'].dt.date
df_weather
```

```python
# Now we need to change the date column in the df, too.

df['timing_of_watering'] = pd.to_datetime(df['timing_of_watering'], format='mixed')
df['timing_of_watering'] = df['timing_of_watering'].dt.date
df
```

```python
# Now we can merge the two dataframes

merged_df = pd.merge(df, df_weather, left_on='timing_of_watering', right_on='date', how='left')
merged_df
```

```python
merged_df.isna().sum()
```

```python
#There are two rows from 2023. We will drop them. And we can drop the date column, because we do not need it in the dataframe twice.
merged_df = merged_df.dropna(subset=['date'])
merged_df = merged_df.drop(['date'], axis=1)
merged_df
```

```python
# We also want to add socio ecnomic values
df_education = pd.read_csv('data_education.csv') 
df_age = pd.read_csv('data_age.csv')
df_income = pd.read_csv('data_income.csv')
df_working_life = pd.read_csv('data_working_life.csv') 
```
